```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

# Personal Approach

The time complexity of this algorithm is O(n). The algorithm splits the input text into two parts: left and right. Then it iterates through each part concatenating words with blank spaces between. While it iterates through each part the lines, the initial word and the final word of each line are stored in tuple. In the end of the right part processing the indexes of the right part are updated and appended to the list of tuples generated by the left part. The output of this algorithm can further be used according to the desired type of alignment. 

```python
columns = ['Index', 'Size', 'Costs', 'Approach', 'Function', 'Time', 'M', 'Memory']
df_LA = pd.read_csv('resultsDVLA_1_8.csv', header=None, names=columns, index_col='Index')
df_RA = pd.read_csv('resultsDVRA_1_8.csv', header=None, names=columns, index_col='Index')
df_CA = pd.read_csv('resultsDVCA_1_8.csv', header=None, names=columns, index_col='Index')
```

As this algorithm after spliting the input text into two it behaves in greedy manner the costs for LA and RA are the same and the costs of CA will be lower.

```python
df_costs_anaylsis = pd.DataFrame([df_LA.Costs, df_RA.Costs, df_CA.Costs]).transpose()
df_costs_anaylsis.columns = ['LA_Costs', 'RA_Costs', 'CA_Costs']
df_costs_anaylsis.iloc[:15,:].plot(y=['LA_Costs', 'RA_Costs', 'CA_Costs'], kind="bar", figsize=(10,9))
```

```python
df_costs_anaylsis.describe()
```

```python
df_memory_anaylsis = pd.DataFrame([df_LA.Memory, df_RA.Memory, df_CA.Memory]).transpose()
df_memory_anaylsis.columns = ['LA_Memory', 'RA_Memory', 'CA_Memory']
df_memory_anaylsis.iloc[:15,:].plot(y=['LA_Memory', 'RA_Memory', 'CA_Memory'], kind="bar", figsize=(10,9))
```

```python
df_memory_anaylsis.describe()
```

The memory consumption does not present significant differences.

```python
df_time_anaylsis = pd.DataFrame([df_LA.Time, df_RA.Time, df_CA.Time]).transpose()
df_time_anaylsis.columns = ['LA_Time', 'RA_Time', 'CA_Time']
df_time_anaylsis.iloc[:15,:].plot(y=['LA_Time', 'RA_Time', 'CA_Time'], kind="bar", figsize=(10,9))
```

```python
df_time_anaylsis.describe()
```

In terms of time of execution the algorithms does not present significant differences neither in memory consumption. For the 54 outputs, the time and memory consumption means are similar.
