# Left Alignment - Analysis

```python
columns = ['Index', 'Size', 'Costs', 'Approach', 'Function', 'Time', 'M', 'Memory']
df_LA = pd.read_csv('resultsGLA_1_8.csv', header=None, names=columns, index_col='Index')
df_LA.head(10)
```

### Analysing the costs and its relation with Ms   & Size

```python
#FIGURE 1
fig, (ax1,ax2, ax3) = plt.subplots(3, figsize=(10,10))
ax1.scatter(df_LA.Costs, df_LA.M)
ax1.set( ylabel='M')
ax2.scatter(df_LA.Costs, df_LA.Size)
ax2.set(ylabel='Size')
ax3.scatter(df_LA.Costs, df_LA.Time)
ax3.set(xlabel='Cost', ylabel='Time')
plt.show()
```

```python
df_LA.corr()
```

### The Time 

```python
#FIGURE 2
fig, (ax1,ax2) = plt.subplots(2, figsize=(10,10))
ax1.scatter(df_LA.Time, df_LA.M)
ax1.set(ylabel='M')
ax2.scatter(df_LA.Time, df_LA.Size)
ax2.set(xlabel='Time', ylabel='Size')
plt.show()
```

## Analysing the top 10 smallest and highest costs.

```python
df_sorted = df_LA.sort_values('Costs')
```

```python
top_10_small_costs = df_sorted.head(10)
```

```python
top_10_high_costs = df_sorted.tail(10)
```

```python
#FIGURE 3
plt.figure(figsize=(8,5))
plt.plot(top_10_small_costs.Costs,top_10_small_costs.Size)
plt.plot(top_10_small_costs.Costs,top_10_small_costs.M)
plt.xlabel("Costs")

plt.legend(loc='lower left',bbox_to_anchor=(0.73, 0.91),
          ncol=2, fancybox=True, shadow=False)
plt.show()
```

```python
#FIGURE 4
plt.figure(figsize=(8,5))
plt.plot(top_10_high_costs.Costs,top_10_high_costs.Size)
plt.xlabel("Costs")
plt.ylabel("Size")
plt.show()
```

The LA algorithm does not present huge differences in terms of costs and time when analysing its outputs in each scenario. It is fair to say that the Ms, the array sizes and the time do not influence strongly the costs (see Figure 1). The memory consumption registered was  constant for each iteration and presented the value of 135966720 bytes. In the Figures  1 and 2 it is clear that there is no linear relation among the parameters mentioned but a small change can be noted when M decreases,  the cost tends to increase.
Considering the top 10 smallest costs computed the parameter M does not influence directly on the costs, neither influences directly the time. In the Figure 3 it is showed how this algorithm behaves with the changes in M and the size of the array. When the array has size of 50 words the cost tend to increase constantly as M varies. However, when the array size differs from 50 the costs follow a pattern of peaks and troughs.
Even though there is no explicit linearity between size and  cost as shown in the Figure 1 when analysing the top 10 highest costs their array sizes are around 150 to 300, after a few oscillations of the cost, then the cost increasement becomes constant when the array size reached 300. 
Briefly, as the size of array increases the time and costs will indeed suffer impacts. However those are going to be really small specially the costs. LA  performance on big arrays might be compared to the performance in small arrays.
